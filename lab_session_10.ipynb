```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

## Stack

```python
class Stack:
    def __init__(self):
        self.head = None

    def push(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def pop(self):
        if self.head is None:
            return None
        popped_data = self.head.data
        self.head = self.head.next
        return popped_data

    def traverse(self):
        curr = self.head
        while curr is not None:
            print(curr.data, end=' ')
            curr = curr.next
```

```python
stack = Stack()
stack.push(10)
stack.push(20)
stack.push(30)
stack.traverse()
```

```python
stack.pop()
stack.pop()
stack.traverse()
```

## Queue

```python
class Queue:
    def __init__(self):
        self.head = None

    def enqueue(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def dequeue(self):
        if self.head is None or self.head.next is None:
            return None
        curr = self.head
        while curr.next.next is not None:
            curr = curr.next
        deleted_element = curr.next.next
        curr.next = None
        return deleted_element

    def traverse(self):
        curr = self.head
        while curr is not None:
            print(curr.data, end=' ')
            curr = curr.next
```

```python
queue = Queue()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
queue.traverse()
```

```python
queue.dequeue()
queue.dequeue()
queue.traverse()
```

## Singly Connected Linked List

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def add(self, data):
        if self.head is None:
            self.head = Node(data)
            return self.head
        curr = self.head
        while curr.next is not None:
            curr = curr.next
        curr.next = Node(data)
        return self.head

    def traverse(self):
        curr = self.head
        while curr is not None:
            print(curr.data, end=' ')
            curr = curr.next

    def search(self, value):
        curr = self.head
        while curr is not None:
            if curr.data is value:
                return f'Value Found'
            curr = curr.next
        return f'Value not Found'

    def insert_before(self, data, reference_data):
        curr = self.head
        new_node = Node(data)
        while curr != None:
            if curr.data == reference_data:
                curr.data, new_node.data = new_node.data, curr.data
                new_node.next, curr.next = curr.next, new_node
                return self.head
            curr = curr.next
```

```python
linkedlist = LinkedList()
linkedlist.add(10)
linkedlist.add(20)
linkedlist.add(30)
linkedlist.traverse()
```

```python
linkedlist.insert_before(25, 30)
linkedlist.traverse()
```

```python
linkedlist.search(30)
```
