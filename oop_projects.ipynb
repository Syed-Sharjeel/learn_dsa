# Hospital Management System
1. Registeration for OPD
2. Entry for Emergency Patients
3. Search for patient record
3. Display record

```python
class Hospital:
    def __init__(self):
        self.opd_patients = []
        self.opd_patients_cnics = []
        self.emergency_patients = []
        self.emergency_patients_cnics = []

    def opd_register(self, name, cnic):
        if len(self.opd_patients) < 250:
            self.opd_patients.append(name)
            self.opd_patients_cnics.append(cnic)
            return f'Thanks for Registeration {name.title()}!'
        else:
            return 'Sorry! No Slot Available Today!'
        
    def emergency_entry(self, name, cnic):
        if len(self.emergency_patients) < 100:
            self.emergency_patients.append(name)
            self.emergency_patients_cnics.append(cnic)
            return f'Entry Successful!'
        else:
            return f'Sorry! No space in emergency ward'

    def search_patient(self, cnic):
        if cnic in self.opd_patients_cnics:
            index = self.opd_patients_cnics.index(cnic)
            return f"Patient Name: {self.opd_patients[index]} CNIC: {cnic}"
        if cnic in self.emergency_patients_cnics:
            index = self.emergency_patients_cnics.index(cnic)
            return f"Patient Name: {self.emergency_patients[index]} CNIC: {cnic}"
        return 'No Record Found'
            
    def display(self, type):
        if type.lower() == 'opd':
            for i in range(len(self.opd_patients)):
                print(f"{i+1}. {self.opd_patients[i]}")
        elif type.lower() == 'emergency':
            for i in range(len(self.emergency_patients)):
                print(f"{i+1}. {self.emergency_patients[i]}")
```

```python
system = Hospital()
system.opd_register('Alice', '12345-666778-1')
```

```python
system.emergency_entry('David', '1111-3322112-2')
```

```python
system.opd_register('Carlos', '11212-111122287-3')
```

```python
system.display('opd')
```

```python
system.display('emergency')
```

```python
system.search_patient('1')
```

```python
system.search_patient('1111-3322112-2')
```

```python
system.search_patient('11212-111122287-3')
```

# Library Management System
1. Add Book
2. Search for Book
3. Display all Books

```python
class Library:
    def __init__(self):
        self.books = []
        self.authors = []
        self.topics = []
    
    def add_book(self, name, author, topic):
        self.books.append(name)
        self.authors.append(author)
        self.topics.append(topic)
        return f'{name} Added to Library'
    
    def search(self, name, author, topic):
        if name:
            if name in self.books:
                return f'{name} is available'
        elif author:
            for i in range(len(self.authors)):
                if self.authors[i] == author:
                    print(f"{self.books[i]}, {self.topics[i]}")
        elif topic:
            for i in range(len(self.topics)):
                if self.topics[i] == topic:
                    print(f"{self.books[i]}, {self.authors[i]}")
    
    def display(self):
        for i in range(len(self.books)):
            print(f"{self.books[i]}\t{self.authors[i]}\t{self.topics[i]}")
```

```python
library_system = Library()
```

```python
library_system.add_book('Fundamentals of Electric Circuits', 'Sadiku', 'Electrical Circuits')
```

```python
library_system.add_book('Calculus 10th edition', 'Howard Anton', 'Calculus')
```

```python
library_system.add_book('Microelectronics Circuit', 'Sedra & Smith', 'Electronic Circuits')
```

```python
library_system.search(name='Calculus 10th edition', author='', topic='')
```

```python
library_system.search(name='', author='', topic='Calculus')
```

# Student Management System
1. Add Student
2. Attendance of Students
3. Add marks for mid terms
4. Add marks for assignment
5. Add marks for finals

```python
class StudentManagement:
    def __init__(self):
        self.names = []
        self.roll_nos = []
        self.attendance_list = []
        self.mids_marks = []
        self.assignment_marks = []
        self.final_marks = []
        self.total_marks = []
    
    def add_student(self, name, roll_no):
        self.names.append(name)
        self.roll_nos.append(roll_no)
        return f'{name} succefully added'
    
    def attendance(self):
        for i in range(len(self.names)):
            attendance = input(f"Attendance for {self.roll_nos[i]}: ")
            self.attendance_list.append(attendance)
        for i in range(len(self.names)):
            print(self.roll_nos[i], self.attendance_list[i])
    
    def add_mids_marks(self):
        for i in range(len(self.names)):
            marks = input(f'Enter Marks for Roll Number {self.roll_nos[i]}')
            self.mids_marks.append(marks)
        for i in range(len(self.names)):
            print(self.roll_nos[i], self.mids_marks[i])
    
    def add_assignment_marks(self):
        for i in range(len(self.names)):
            marks = input(f'Enter Marks for Roll Number {self.roll_nos[i]}')
            self.assignment_marks.append(marks)
        for i in range(len(self.names)):
            print(self.roll_nos[i], self.assignment_marks[i])
    
    def add_final_marks(self):
        for i in range(len(self.names)):
            marks = input(f'Enter Marks for Roll Number {self.roll_nos[i]}')
            self.final_marks.append(marks)
            self.total_marks.append(self.mids_marks[i] + self.assignment_marks[i] + self.final_marks[i])
            
        for i in range(len(self.names)):
            print(self.roll_nos[i], self.final_marks[i])
```

```python
student_system = StudentManagement()
```

```python
student_system.add_student('Carlos', 1)
```
