# Lab Session 04
_Comparision between sorting algorithms_
- Selection Sort
- Bubble Sort
- Insertion Sort
- Merge Sort

```python
import time
import matplotlib.pyplot as plt
```

## Selection Sort

```python
def selection_sort(a):
  for i in range(len(a)):
    min = i
    j = i + 1
    for j in range(len(a)):
      if a[min] < a[j]:
        min = j
      a[i], a[min] = a[min], a[i]
  return a
```

# Bubble Sort

```python
def bubble_sort(a):
    n = len(a)
    for i in range(1,n):
        for j in range(n-i):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
    return a
```

# Insertion Sort

```python
def insertion_sort(a):
    for j in range(1, len(a)):
        key = a[j]
        i = j - 1
        while i >= 0 and a[i] > key:
            a[i+1] = a[i]
            i -= 1
        a[i+1] = key
    return a
```

# Merge Sort

```python
def merge_sort(A):
   n=len(A)
   s=list()
   if n==1:
      s=A
   else:
      a=(n//2)
      s1=merge_sort(A[0:a])
      s2=merge_sort(A[a:n])
      s=merge(s1,s2)
   return s
def merge(A,B):
    n1=len(A)
    n2=len(B)
    A=A+[float('inf')]
    B=B+[float('inf')]
    i=0
    j=0
    M=[]
    for k in range(n1+n2):
        if A[i] < B[j]:
            M.append( A[i] )
            i=i+1
        else:
            M.append ( B[j] )
            j=j+1
    return M
```

### Verification

```python
print(selection_sort([5,4,3,2,1]))
print(bubble_sort([5,4,3,2,1]))
print(insertion_sort([5,4,3,2,1]))
print(merge_sort([5,4,3,2,1]))
```

# Runtime Calculation

## Selection Sort

```python
sizes = []
b_t_selection = []
w_t_selection = []

for x in range(2000, 24000, 2000):
    sizes.append(x)

    # BEST CASE
    best_array = list(range(0, x))
    ti = time.time()
    selection_sort(best_array)
    tf = time.time()
    b_t_selection.append(tf - ti)

    # WORST CASE
    worst_array = list(range(x, 0, -1))
    ti = time.time()
    selection_sort(worst_array)
    tf = time.time()
    w_t_selection.append(tf - ti)
```

## Bubble Sort

```python
sizes = []
b_t_bubble = []
w_t_bubble = []

for x in range(2000, 24000, 2000):
    sizes.append(x)

    # BEST CASE
    best_array = list(range(0, x))
    ti = time.time()
    bubble_sort(best_array)
    tf = time.time()
    b_t_bubble.append(tf - ti)

    # WORST CASE
    worst_array = list(range(x, 0, -1))
    ti = time.time()
    bubble_sort(worst_array)
    tf = time.time()
    w_t_bubble.append(tf - ti)
```

## Insertion Sort

```python
sizes = []
b_t_insertion = []
w_t_insertion = []

for x in range(2000, 24000, 2000):
    sizes.append(x)

    # BEST CASE
    best_array = list(range(0, x))
    ti = time.time()
    insertion_sort(best_array)
    tf = time.time()
    b_t_insertion.append(tf - ti)

    # WORST CASE
    worst_array = list(range(x, 0, -1))
    ti = time.time()
    insertion_sort(worst_array)
    tf = time.time()
    w_t_insertion.append(tf - ti)
```

# Merge Sort

```python
b_t_mergesort = []
w_t_mergesort = []
sizes = []

for i in range(2000,24000,2000):
    best_array = list(range(0,i,1))
    worst_array = list(range(i,0,-1))
    sizes.append(i)

    # BEST TIME ANALYSIS
    ti = time.time()
    merge_sort(best_array)
    tf = time.time()
    b_t_mergesort.append(tf - ti)

    # WORST TIME ANALYSIS
    ti = time.time()
    merge_sort(worst_array)
    tf = time.time()
    w_t_mergesort.append(tf - ti)
```

## Comparision for Best Cases

```python
plt.title(f'Comparison of Best Case for Selection Sort, Bubble Sort, and Insertion Sort Algorithms')
plt.xlabel('Input sizes')
plt.ylabel('Time (sec)')
plt.plot(sizes, b_t_selection, color='blue', label='Selection Sort')
plt.plot(sizes, b_t_bubble, color='red', label='Bubble Sort')
plt.plot(sizes, b_t_insertion, color='green', label='Insertion Sort')
plt.plot(sizes, b_t_mergesort, color='orange', label='Merge Sort')
plt.legend()
plt.grid()
plt.show()
```

## Comparision for Worst Cases

```python
plt.title(f'Comparison of Worst Case for Selection Sort, Bubble Sort, and Insertion Sort Algorithms')
plt.xlabel('Input sizes')
plt.ylabel('Time (sec)')
plt.plot(sizes, w_t_selection, color='blue', label='Selection Sort')
plt.plot(sizes, w_t_bubble, color='red', label='Bubble Sort')
plt.plot(sizes, w_t_insertion, color='green', label='Insertion Sort')
plt.plot(sizes, w_t_mergesort, color='orange', label='Merge Sort')
plt.legend()
plt.grid()
plt.show()
```
