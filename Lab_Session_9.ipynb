# Stack

```python
class Stack:
    def __init__(self):
        self.stack = []
    def push(self, data):
        self.stack.insert(0, data)
    def pop(self):
        self.stack.remove(self.stack[0])
```

```python
stack = Stack()
stack.push(10)
stack.push(500)
stack.push(-5)
```

```python
stack.pop()
```

```python
print(stack.stack)
```

# Queue

```python
class Queue:
    def __init__(self):
        self.queue = []
    def enqueue(self, data):
        self.queue.insert(0, data)
    def dequeue(self):
        self.queue.remove(self.queue[-1])
```

```python
queue = Queue()
queue.enqueue(10)
queue.enqueue(500)
queue.enqueue(-5)
```

```python
queue.dequeue()
```

```python
print(queue.queue)
```

# Dynamic Set

```python
class DynamicSet:
    def __init__(self):
        self.dynamic_set = []
    def search(self, data):
        if data in self.dynamic_set:
            return True
        else:
            return False
    def insert(self, data):
        self.dynamic_set.append(data)
    def delete(self, data):
        self.dynamic_set.remove(data)
    def minimum(self):
        print(min(self.dynamic_set))
    def maximum(self):
        print(max(self.dynamic_set))
```

```python
dynamic_set = DynamicSet()
dynamic_set.insert(10)
dynamic_set.insert(500)
dynamic_set.insert(-250)
```

```python
dynamic_set.minimum()
```

```python
dynamic_set.maximum()
```

```python
dynamic_set.search(500)
```

```python
dynamic_set.search(-1000)
```

```python
dynamic_set.delete(500)
```
