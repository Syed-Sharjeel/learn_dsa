```python
import time
import matplotlib.pyplot as plt
```

1. Implement Linear Search Algorithm.
2. Create a function named Linear Search that takes a list and target element as its input arguments.
3. Estimate the total running time of Linear Search for the b_t and w_t Cases.

```python
def linear_search(x, key):
    count = 0
    flag = 0
    for i in range(len(x)):
        count += 1
        if x[i] == key:
            flag = 1
    return flag
```

```python
linear_search([5,4,3,2,1], 5)
```

```python
sizes = []
ls_b_t = []
ls_w_t = []
for i in range(1000, 100000, 500):
    array = list(range(i))
    sizes.append(len(array))

    t_i = time.time()
    linear_search(array, 0)
    t_f = time.time()
    ls_b_t.append(t_f - t_i)

    t_i = time.time()
    linear_search(array, -1)
    t_f = time.time()
    ls_w_t.append(t_f - t_i)
```

```python
plt.figure(figsizes=(10,8))
plt.plot(sizes, ls_b_t, color='orange', label='Linear b_t Case')
plt.plot(sizes, ls_w_t, color='green', label='Linear w_t Case')
plt.title('b_t & w_t Case Analysis for Linear Search Algorithm')
plt.xlabel('sizes of Array')
plt.ylabel('Time')
plt.grid()
plt.legend()
plt.show()
```

1. Implement the Binary Search Algorithm.
2. Create a function named Binary Search that takes a list and target element as its input arguments.
3. Estimate the total running time of Binary Search for the b_t and w_t *Cases* for YOUR implementation and plot a growth plot.

```python
def binary_search(A, key):
    f_index = 0
    l_index = len(A) - 1
    flag = 0

    while f_index <= l_index and flag == 0:
        mid = (f_index + l_index) //2
        if A[mid] == key:
            flag = 1
        elif A[mid] > key:
            l_index = mid - 1
        else:
            f_index = mid + 1
    return flag
```

```python
binary_search([1,2,3,4,5],5)
```

```python
bs_b_t = []
bs_w_t = []

for i in range(1000, 100000, 500):
    array = list(range(i))
    t_i = time.time()
    mid = ((0 + len(array)) // 2)
    binary_search(array, array[mid])
    t_f = time.time()
    bs_b_t.append(t_f - t_i)

    t_i = time.time()
    binary_search(array, -1)
    t_f = time.time()
    bs_w_t.append(t_f - t_i)
```

```python
plt.figure(figsizes=(10,5))
plt.plot(sizes, bs_b_t, color='orange', label='Binary b_t Case')
plt.plot(sizes, bs_w_t, color='green', label='Binary w_t Case')
plt.title('b_t & w_t Case Analysis for Binary Search Algorithm')
plt.xlabel('sizes of Array')
plt.ylabel('Time')
plt.grid()
plt.legend()
plt.show()
```
