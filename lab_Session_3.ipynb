```python
import time
import matplotlib.pyplot as plt
```

# Selection Sort

```python
def selection_sort(a):
  for i in range(len(a)):
    min = i
    j = i + 1
    for j in range(len(a)):
      if a[min] < a[j]:
        min = j
      a[i], a[min] = a[min], a[i]
  return a
```

```python
selection_sort([5,4,3,2,1])
```

```python
size = []
ss_t_best = []
ss_t_worst = []

for x in range(2000, 24000, 2000):
    size.append(x)

    # BEST CASE
    best_array = list(range(0, x))
    init_time = time.time()
    selection_sort(best_array)
    fin_time = time.time()
    ss_t_best.append(fin_time - init_time)

    # WORST CASE
    worst_array = list(range(x, 0, -1))
    init_time = time.time()
    selection_sort(worst_array)
    fin_time = time.time()
    ss_t_worst.append(fin_time - init_time)
```

# Bubble Sort

```python
def bubble_sort(a):
  n = len(a)
  for i in range(1,n):
    for j in range(n-i):
      if a[j] > a[j+1]:
        a[j], a[j+1] = a[j+1], a[j]
  return a
```

```python
bubble_sort([5,4,3,2,1])
```

```python
size = []
bs_t_best = []
bs_t_worst = []

for x in range(2000, 24000, 2000):
    size.append(x)

    # BEST CASE
    best_array = list(range(0, x))
    init_time = time.time()
    bubble_sort(best_array)
    fin_time = time.time()
    bs_t_best.append(fin_time - init_time)

    # WORST CASE
    worst_array = list(range(x, 0, -1))
    init_time = time.time()
    bubble_sort(worst_array)
    fin_time = time.time()
    bs_t_worst.append(fin_time - init_time)
```

# Insertion Sort

```python
def insertion_sort(a):
  for j in range(1, len(a)):
    key = a[j]
    i = j - 1
    while i >= 0 and a[i] > key:
      a[i+1] = a[i]
      i -= 1
    a[i+1] = key
  return a
```

```python
insertion_sort([5,4,3,2,1])
```

```python
size = []
is_t_best = []
is_t_worst = []

for x in range(2000, 24000, 2000):
    size.append(x)

    # BEST CASE
    best_array = list(range(0, x))
    init_time = time.time()
    insertion_sort(best_array)
    fin_time = time.time()
    is_t_best.append(fin_time - init_time)

    # WORST CASE
    worst_array = list(range(x, 0, -1))
    init_time = time.time()
    insertion_sort(worst_array)
    fin_time = time.time()
    is_t_worst.append(fin_time - init_time)
```

# Plots

```python
plt.title(f'Comparison of Best Case for Selection Sort, Bubble Sort, and Insertion Sort Algorithms')
plt.xlabel('Input Size')
plt.ylabel('Time (sec)')
plt.plot(size, ss_t_best, color='blue', label='Selection Sort')
plt.plot(size, bs_t_best, color='red', label='Bubble Sort')
plt.plot(size, is_t_best, color='green', label='Insertion Sort')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.title(f'Comparison of Worst Case for Selection Sort, Bubble Sort, and Insertion Sort Algorithms')
plt.xlabel('Input Size')
plt.ylabel('Time (sec)')
plt.plot(size, ss_t_worst, color='blue', label='Selection Sort')
plt.plot(size, bs_t_worst, color='red', label='Bubble Sort')
plt.plot(size, is_t_worst, color='green', label='Insertion Sort')
plt.legend()
plt.grid()
plt.show()
```
